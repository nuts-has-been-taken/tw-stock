### 日元對美元匯率

接下來觀察匯率對市場影響

```python
import yfinance as yf
import matplotlib.pyplot as plt

start='2000-01-01'
end='2024-08-23'

jpy_usd = yf.download('JPY=X', start=start, end=end)
sp500 = yf.download('^GSPC', start=start, end=end)

jpy_usd['Normalized Close'] = (jpy_usd['Close'] - jpy_usd['Close'].min()) / (jpy_usd['Close'].max() - jpy_usd['Close'].min())
sp500['Normalized Close'] = (sp500['Close'] - sp500['Close'].min()) / (sp500['Close'].max() - sp500['Close'].min())

plt.figure(figsize=(12, 6))
plt.plot(sp500.index, sp500['Normalized Close'], label='S&P 500 Close Normalized', color='orange')
plt.plot(jpy_usd.index, jpy_usd['Normalized Close'], label='USD/JPY Close Normalized', color='purple')
plt.xlabel('Date')
plt.ylabel('Normalized Growth Rate (%)')
plt.grid(True)
plt.legend()
plt.show()
```

匯率對市場影響看起來沒有很明顯，反而看起來像是市場的落後指標
市場價格的下行反應在匯率的滯後性並且持續一段時間
